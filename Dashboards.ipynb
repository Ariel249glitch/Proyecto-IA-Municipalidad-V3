```python
#Consultas y Errores

import matplotlib.pyplot as plt

consultas = 5
errores = 0

plt.figure(figsize=(6,4))
plt.bar(["Consultas", "Errores"], [consultas, errores])

plt.title("Consultas Procesadas y Errores")
plt.ylabel("Cantidad")
plt.show()
```

```python
#latencia

import matplotlib.pyplot as plt

latencia = 3.82

plt.figure(figsize=(6,4))
plt.bar(["Latencia"], [latencia])

plt.title("Latencia Promedio del Sistema")
plt.ylabel("Segundos")
plt.show()
```

```python
#Distrbucion y consultas

import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.pie(
    [5, 0.0001],
    labels=["Exitosas", "Errores"],
    autopct="%1.1f%%"
)

plt.title("Distribución de Consultas")
plt.show()
```
